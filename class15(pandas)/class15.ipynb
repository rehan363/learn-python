# pandas core components
* series types
*  dataframe types

```python
# List we can use for creating series

s1 : pd.Series = pd.Series((1, 2, 3, 4, 5))
s1
```

```python
# dictionary we can use for creating series

s1 : pd.Series = pd.Series({"a":80, 
                            "b":90,
                            "c":100, 
                            "d":110, 
                            "e":120})
s1
```

```python
values : list[int] = [0,1,2,3,4,5]
index1 : list[list[str]] = [['a1', 'a1','a2','a2','a3','a3'],
                            ['a', 'b', 'c', 'd', 'e','f']]

s1 : pd.Series = pd.Series (values , index=index1)
s1
```

```python
values : list[int] = [0,1,2,3,4,5]
index1 : list[list[str]] = [['a1', 'a1','a2','a2','a3','a3'],
                            ['a', 'b', 'c', 'd', 'e','f']]

s1 : pd.Series = pd.Series (values , index=index1 , name="student_data")
s1
```

# dataframe

```python
s1 :pd.Series = pd.Series ([1, 2, 3, 4, 5], name = "student_id")
s2 :pd.Series = pd.Series([10, 20, 30, 40, 79] , name = "score")
s3 :pd.Series = pd.Series(["rehan", "ali", "ahmed", "hamza", "zia"] , name = "student_name")

df1 : pd.DataFrame = pd.DataFrame({"student_id":s1, "score":s2 , "student_name":s3})

df1
```

ANOTHER METHOOD TO COMBINE DATA

```python
s1 :pd.Series = pd.Series ([1, 2, 3, 4, 5], name = "student_id")
s2 :pd.Series = pd.Series([10, 20, 30, 40, 79] , name = "score")
s3 :pd.Series = pd.Series(["rehan", "ali", "ahmed", "hamza", "zia"] , name = "student_name")

df1 : pd.DataFrame = pd.concat([s1, s2, s3], axis=1)

df1
```

```python
Data : list[list[int]] = [[1, 2, 3],
                         [4, 5, 6],
                         [7, 8, 9]]

df1 : pd.DataFrame = pd.DataFrame (Data)
df1
```

TO DEFINE columns

```python
Data : list[list[int]] = [[1, 2, 3],
                         [4, 5, 6],
                         [7, 8, 9]]

df1 : pd.DataFrame = pd.DataFrame (Data, columns=["A", "B", "C"])
df1
```

TO DEFINE INDEX 

```python
Data : list[list[int]] = [[1, 2, 3],
                         [4, 5, 6],
                         [7, 8, 9]]

df1 : pd.DataFrame = pd.DataFrame (Data, columns=["A", "B", "C"], index=["X", "Y", "Z"])
df1
```
